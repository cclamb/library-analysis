```python
from elftools.elf.elffile import ELFFile
from elftools.elf.relocation import RelocationSection
from elftools.elf.sections import SymbolTableSection

f = open('data/sample_exe64.elf', 'rb')
elf_file = ELFFile(f)

dwarf_info =  elf_file.get_dwarf_info()

print 'Sections...'
print 'Looks like we have %s' % elf_file['e_shnum']
for section in elf_file.iter_sections():
    if section.name == '.rela.dyn':
        print '\t-> relocation section %s' % section.name
        print '\t\t%s section with %s relocation(s)' % (section.name, section.num_relocations())
        
        for relocation in section.iter_relocations():
            print '\t\tRelocation (%s)' % 'RELA' if relocation.is_RELA() else 'REL'
            print '\t\toffset = %s' % relocation['r_offset']
    if section.name == '.symtab':
        header = section.header
        print '\t-> symbol table %s' % section.name
        print '\t\tSection name: %s, type: %s' % (header['sh_name'], header['sh_type'])
        print '\t\tIt is a symbol section with %s symbols' % section.num_symbols()
        for symbol in section.iter_symbols():
            print '\t\t\tname: %s entry: %s' % (symbol.name, symbol.entry)
    else:
        print '\tname=%s' % section.name

print 'Compilation Units...'
for CU in dwarf_info.iter_CUs():
    print 'DIR on CU: '
    print 'Found a compile unit at offset %s, length %s' % (CU.cu_offset, CU['unit_length'])
    top_DIE = CU.get_top_DIE()
    print '\tTop DIE with tag=%s' % top_DIE.tag
    print '\tname=%s' % top_DIE.get_full_path()
    print '\taddress_size is %s' % CU['address_size']
    
print type(elf_file.elfclass)
```
